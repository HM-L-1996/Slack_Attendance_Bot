```python
import pandas as pd
import numpy as np
df = pd.read_excel(r'C:\Users\Playdata\OneDrive\miniPrj04.4.xlsx', engine='openpyxl')
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df= df.drop([0])

#배열로 입력해야 하는 부분입니다.
date = '2022.03.01'
index = 1
checkResult = ['○','○','지각','○','○','○','○','지각','○','○','○','지각','○','○','○','○','○','○','○',np.nan,'○','○','○','○','○',np.nan,'○','○',np.nan]


#배열의 요소들 dataframe(df)에 입력
df.insert(index,date,checkResult)
#오늘 참석자 수 date
df.at[29, date] = df.notnull().sum(axis=0)[index]
#df[date] = df.apply(lambda row: sum(row[:].notnull()) ,axis=1)
#df[date] = df[df != None].count(axis=1)

#참석일 attend_days
df['attend_days'] = df[df == '○'].count(axis=1)
df.at[29, 'attend_days'] = df['attend_days'].sum()

#지각일 late_days
df['late_days'] = df.apply(lambda row: sum(row[:]=='지각') ,axis=1)
df.at[29, 'late_days'] = df['late_days'].sum()

#결석일 absent_days
#df['absent_days'] = df.isnull().sum(axis = 1)
#df.at[29, 'absent_days'] = df['absent_days'].sum()
#df['absent_days'] = df[df == None].count(axis=1)
df['absent_days'] = df.apply(lambda row: sum(row[:].isnull()) ,axis=1)
#df.at[29, 'absent_days'] = df['absent_days'].sum()

#출석인정일 sum 
df['sum'] = df['attend_days']-df['absent_days']-np.trunc(df['late_days']/3)
#df.at[29, 'sum'] = df['sum'].sum()

df#출력 확인
#df.to_excel(r'C:\Users\Playdata\OneDrive\miniPrj04.4.xlsx', index=False) #OneDrive에 파일 덮어쓰기 가능 
```
